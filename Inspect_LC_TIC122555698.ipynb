```python
import numpy as np
import matplotlib.pyplot as plt
import lightkurve

infile3 = 'tess2018263035959-s0003-0000000122555698-0123-s_tp.fits'
infile4 = 'tess2018292075959-s0004-0000000122555698-0124-s_tp.fits'

sector3 = lightkurve.open(infile3)
sector4 = lightkurve.open(infile4)
```

Let's have a quick look at the shape of things:

```python
sector3.plot(frame=0)
sector4.plot(frame=0)
```

```python
lc_ap3 = sector3.to_lightcurve(aperture_mask='pipeline').remove_outliers()
lc_ap4 = sector4.to_lightcurve(aperture_mask='pipeline').remove_outliers()

print("Sector 3 :", np.median(lc_ap3.flux),",", np.mean(lc_ap3.flux))
print("Sector 4 :", np.median(lc_ap4.flux),",", np.mean(lc_ap4.flux))
```

```python
print(lc_ap4.time[7000], sector4.time[7000])
```

```python
lc_ap3.plot(normalize=False)
lc_ap4.plot(normalize=False)
```

```python
pipelc3 = lightkurve.open('tess2018263035959-s0003-0000000122555698-0123-s_lc.fits')
pipelc3.plot(normalize=False)
pipelc4 = lightkurve.open('tess2018292075959-s0004-0000000122555698-0124-s_lc.fits')
pipelc4.plot(normalize=False)
```

```python
print(np.nanmedian(pipelc3.SAP_FLUX.flux) / np.nanmedian(pipelc3.PDCSAP_FLUX.flux))
print(np.nanmedian(pipelc4.SAP_FLUX.flux) / np.nanmedian(pipelc4.PDCSAP_FLUX.flux))
```
